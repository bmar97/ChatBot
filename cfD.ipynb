### (1) Capital flows in developing economies over time 

```python
# Define Panel widgets
year_slider = pn.widgets.IntSlider(name='Year slider', start=1982, end=2023)
```

```python
# Radio buttons for CO2 measures
yaxis_GV = pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['total_victims', 'fatalities', 'injured', 'age_of_shooter'],
    button_type='success'
)
```

```python
df.groupby(['location']).count().sort_values(by='year', ascending=False).head(5)
```

```python
# Chart plot of 5 largest states
states = ['California', 'Texas', 'Florida', 'Colorado',' Washington']
GV_pipeline = (
    idf[
        (idf.year <= year_slider) &
        (idf.location.isin(states))
    ]
    .groupby(['location', 'year'])[yaxis_GV].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
GV_plot = GV_pipeline.hvplot(x ='year', by='location', y=yaxis_GV, line_width=2, title="Gun Violence")
GV_plot
```

### (2) Table - GDP over time by emerging enconomy

```python
GV_table = GV_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
#GV_table
```

```python
states = ['California', 'Texas', 'Florida', 'Colorado',' Washington']

yaxis_GV_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['total_victims', 'fatalities', 'injured', 'age_of_shooter'], 
    button_type='success'
)

GV_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.location.isin(states))
    ]
    .groupby(['year', 'location'])[yaxis_GV_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```
